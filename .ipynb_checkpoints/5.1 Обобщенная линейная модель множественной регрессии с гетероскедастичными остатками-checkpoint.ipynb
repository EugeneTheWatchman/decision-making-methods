```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ввод данных из примера
data = {
    'xi': [3, 6, 7, 9, 13, 15, 18, 21, 22, 24, 25, 26, 27, 28, 35, 37, 44],
    'yi': [18, 27, 18, 45, 55, 68, 51, 84, 85, 100, 63, 130, 135, 60, 70, 80, 180]
}

# Создаем DataFrame
df = pd.DataFrame(data)

# Построение модели линейной регрессии
model = ols('yi ~ xi', data=df).fit()

# Вывод результатов
print(model.summary())
```

1.  **Общие характеристики модели**:
    
    *   **R-квадрат**: 0.608. Это означает, что около 60.8% изменчивости зависимой переменной yi объясняется моделью с помощью независимой переменной xi.
    *   **Скорректированный R-квадрат**: 0.582, что также указывает на адекватность модели.
    *   **F-статистика**: 23.31 с p-значением 0.000221. Это свидетельствует о статистической значимости модели в целом на уровне значимости 0.05.
2.  **Дополнительные тесты**:
    
    *   **Durbin-Watson**: 1.744, что указывает на отсутствие существенной автокорреляции в остатках модели.
    *   **Тесты на нормальность распределения остатков** (Omnibus, Jarque-Bera) показывают, что остатки достаточно близки к нормальному распределению.

*   Модель адекватно описывает зависимость между yi и xi, поскольку коэффициент при xi статистически значим и объясняет значительную часть вариации yi.
*   Несмотря на то что константа не является статистически значимой, она играет роль в модели, представляя собой ожидаемое значение yi при xi равном 0.
*   Высокий коэффициент детерминации указывает на хорошую подгонку модели к данным.​​

# Пункт 2

```python
import matplotlib.pyplot as plt
import numpy as np

# Построение графика зависимостей
plt.figure(figsize=(12, 6))

# График наблюдаемых значений
plt.subplot(1, 2, 1)
plt.scatter(df['xi'], df['yi'], color='blue', label='Наблюдаемые значения')
plt.plot(df['xi'], model.fittedvalues, color='red', label='Линия регрессии')
plt.title('График зависимости yi от xi')
plt.xlabel('Численность работающих (xi, тыс. чел.)')
plt.ylabel('Доходы (yi, млрд. руб.)')
plt.legend()

# График остатков
plt.subplot(1, 2, 2)
plt.scatter(df['xi'], model.resid, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('График остатков регрессии')
plt.xlabel('Численность работающих (xi, тыс. чел.)')
plt.ylabel('Остатки')

plt.tight_layout()
plt.show()
```

1. **На графике зависимости** видно, что линия регрессии проходит через центральные точки данных, что говорит о том, что модель обладает определенной степенью точности. Однако есть некоторые точки, которые существенно отклоняются от линии регрессии, что может указывать на возможные выбросы или наличие других переменных, влияющих на yi.

2. **На графике остатков** идеальная модель должна показывать случайное распределение остатков вокруг горизонтальной линии нуля без явных паттернов. Если остатки расположены неслучайным образом, это может указывать на проблемы в модели, такие как гетероскедастичность или неправильная спецификация модели.

```python
from scipy.stats import spearmanr

# Вычисление абсолютных значений остатков
df['abs_resid'] = abs(model.resid)

# Ранжирование значений xi и абсолютных остатков
df['rank_xi'] = df['xi'].rank()
df['rank_abs_resid'] = df['abs_resid'].rank()

# Вычисление коэффициента ранговой корреляции Спирмена
spearman_corr, spearman_p_value = spearmanr(df['rank_xi'], df['rank_abs_resid'])
spearman_corr, spearman_p_value
```

Результаты теста ранговой корреляции Спирмена, которые вы получили, указывают на следующее:

1.  **Коэффициент корреляции Спирмена** составляет 0.8652. Это высокое положительное значение коэффициента указывает на сильную монотонную зависимость между рангами независимой переменной xi и абсолютными значениями остатков. В контексте анализа регрессии это может быть интерпретировано как признак потенциальной гетероскедастичности, поскольку предполагает, что с увеличением значений xi увеличивается и разброс остатков.
    
2.  **P-значение** равно примерно 7.32×10^−6, что значительно меньше стандартного уровня значимости 0.05. Это говорит о том, что наблюдаемая корреляция между рангами xi и абсолютными остатками статистически значима и вряд ли является случайной.

```python
from statsmodels.stats.diagnostic import het_goldfeldquandt

# Выполнение теста Голдфелда-Квандта
gq_test = het_goldfeldquandt(df['yi'], df[['xi']], split=0.5, drop=0)
gq_test_statistic, gq_test_p_value, _ = gq_test

gq_test_statistic, gq_test_p_value
```

* Статистика теста значительно выше 1, что указывает на то, что дисперсия остатков в одной части данных выше, чем в другой. Это свидетельствует в пользу наличия гетероскедастичности.
* P-значение намного меньше уровня значимости
α=0.05, следовательно, нулевая гипотеза о гомоскедастичности (равенстве дисперсий остатков в разных подгруппах) отвергается. Это подтверждает наличие гетероскедастичности в данных.

```python
df['abs_resid'] = abs(model.resid)

# Значения параметра δ для тестирования
delta_values = [1/3, 1/2, 1, 2, 3]
f_statistics = {}

for delta in delta_values:
    # Преобразование xi в соответствии с δ
    df[f'xi_transformed_{str(delta).replace(".", "_")}'] = df['xi'] ** delta

    # Построение новой регрессионной модели
    formula = f'abs_resid ~ xi_transformed_{str(delta).replace(".", "_")}'
    transformed_model = ols(formula, data=df).fit()

    # Сохранение F-статистики
    f_statistics[delta] = transformed_model.fvalue

# Нахождение δ с наибольшей F-статистикой
best_delta = max(f_statistics, key=f_statistics.get)
best_f_statistic = f_statistics[best_delta]

best_delta, best_f_statistic
```

На основе выполненного тестирования абсолютных остатков методом Глейзера с различными значениями параметра δ вы получили следующие результаты:

*   Лучшее значение параметра δ, для которого величина F-статистики максимальна, равно 1.
*   Соответствующее наибольшее значение F-статистики составляет 34.6788.

*   Параметр δ, равный 1, оказался наиболее подходящим для преобразования переменной xi в вашей модели. Это указывает на то, что преобразование переменной в её текущем виде (без изменения) наиболее эффективно с точки зрения объяснения вариативности абсолютных остатков модели.
*   Высокое значение F-статистики для δ=1 свидетельствует о значимом улучшении модели по сравнению с моделями с другими значениями δ, особенно в контексте учета гетероскедастичности.
*   Эти результаты подтверждают, что дальнейшее использование модели без преобразования xi является предпочтительным вариантом для минимизации гетероскедастичности остатков и улучшения качества модели.


```python
df['weights'] = 1 / df['xi'] ** 2

# Построение модели ВНК
wls_model = sm.WLS(df['yi'], sm.add_constant(df['xi']), weights=df['weights']).fit()

# Вывод результатов
print(wls_model.summary())
```

## Выводы по результатам ВНК:

- **Улучшение модели**: Модель ВНК показала значительное улучшение в объяснительной способности по сравнению с OLS, с коэффициентом детерминации (R-квадрат) на уровне **0.828**, что указывает на то, что около **82.8%** вариации yi может быть объяснено моделью. Скорректированный R-квадрат **0.817** подтверждает хорошее соответствие модели данным.

- **Статистическая значимость коэффициентов**: Оба коэффициента модели - константа и коэффициент при xi - являются статистически значимыми, с p-значениями **0.041** и **<0.001** соответственно, подтверждая их важность в модели.

- **F-статистика**: Высокая F-статистика в **72.46** с p-значением **3.98e-07** демонстрирует статистическую значимость модели в целом на уровне значимости **0.05**.

- **Коррекция гетероскедастичности**: Применение весов в ВНК корректно учло гетероскедастичность, повысив тем самым надежность стандартных ошибок и статистических выводов.

- **Заключение**: Результаты указывают на адекватность использования ВНК для анализа данных с гетероскедастичностью остатков, обеспечивая более надежные и точные оценки по сравнению с уравнение линейной регрессии.
