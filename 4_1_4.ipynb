```python
import pandas as pd

data_frame = pd.read_csv('/content/test_data_3.csv', delimiter=';', encoding='windows-1251')
data_frame
```

```python
import matplotlib.pyplot as plt

# Since the data in 'Потребление электроэнергии, yt' column uses comma as a decimal separator, we need to convert it to float
data_frame['Потребление электроэнергии, yt'] = data_frame['Потребление электроэнергии, yt'].str.replace(',', '.').astype(float)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(data_frame['Номер квартала, t'], data_frame['Потребление электроэнергии, yt'], marker='o')
plt.title('График зависимости потребления электроэнергии от номера квартала')
plt.xlabel('Номер квартала, t')
plt.ylabel('Потребление электроэнергии, yt')
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

# Plotting the autocorrelation plot (Кореллограмма)
plt.figure(figsize=(10, 6))
plot_acf(data_frame['Потребление электроэнергии, yt'].dropna(), lags=8)
plt.title('Кореллограмма временного ряда потребления электроэнергии')
plt.xlabel('Лаг')
plt.ylabel('Автокорреляция')
plt.show()
```
