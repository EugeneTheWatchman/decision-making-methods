```python
import pandas as pd
```

```python
columns = [
  'Цена, руб',
  'Мощность двигателя, Вт',
  'Максимальная скорость, км/ч',
  'Запас хода, км',
  'Диаметр колес, см',
  'Максимальная нагрузка, кг',
  'Максимальный угол подъема, °',
  'Вес, кг',
  'Время зарядки, ч',
]

index = [
    'MEKOTRON Hoverboard 6.5',
    'HOVERBOT B4 Premium',
    'ICONBIT Smart',
    'HOVERBOT A-15 Premium',
    'BESSHOF AJPY6-1',
    'BESSHOF AJPY10-1',
    'MEKOTRON Hoverboard 6.6',
]
data = [
    [9990, 250, 15, 20, 16.51, 100, 15, 10.2, 3],
    [19990, 350, 15, 15, 20.32, 120, 10, 12, 2],
    [8390, 700, 15, 20, 16.51, 110, 15, 10.5, 2],
    [17990, 400, 15, 15, 16.51, 120, 10, 10.5, 2],
    [8990, 250, 10, 10, 16.51, 100, 10, 10.2, 1.5],
    [12990, 250, 10, 10, 25.4, 100, 10, 12.4, 1.5],
    [12990, 250, 15, 20, 16.51, 100, 15, 10.2, 3],
]

df = pd.DataFrame(data, columns=columns, index=index)
df.index.name = 'Модель'
df
```

```python
df_normalized = df.copy()
for col in df_normalized.columns:
    if col in ["Цена, руб", "Вес, кг", "Время зарядки, ч"]:
        df_normalized[col] = 1 / df_normalized[col]
    df_normalized[col] = df_normalized[col] / df_normalized[col].max()

df_normalized
```
